```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create graph for aroma at different levels of function length vs similarity requirements

df = pd.read_csv("outputAdditional.csv")
# print(df)

df['recall'] = df.apply(lambda x: x['foundAroma'] / x['potential'], axis=1)

df['unixcoderFoundPercentage'] = df.apply(lambda x: x['foundUnixcoder'] / x['potential'], axis=1)
# print(df['foundPercentage'])

dropAmounts = [0, 0.25, 0.5, 0.75, 0.9]

for val in dropAmounts:
    dfDrop = df[df.dropAmount == val]
    plt.plot(dfDrop['sim'], dfDrop['recall'], label="amount dropped: " + str(val))
plt.xlabel("similarity cutoff")
plt.ylabel("recall")
plt.legend()
plt.show()






for val in dropAmounts:
    dfDrop = df[df.dropAmount == val]
    plt.plot(dfDrop['sim'], dfDrop['unixcoderFoundPercentage'], label="amount dropped: " + str(val))
plt.xlabel("similarity cutoff")
plt.ylabel("recall")
plt.legend()
plt.show()
    
```

```python
# precision vs recall
df['precision'] = df.apply(lambda x: x['foundAroma'] / (x['foundAroma'] + x['unrelatedAroma']), axis=1)
for val in dropAmounts:
    dfDrop = df[df.dropAmount == val]
    plt.plot(dfDrop['recall'], dfDrop['precision'], label="amount dropped: " + str(val))

print(max(df['recall'] * df['precision'] * 2 / (df['recall'] + df['precision'])))
plt.xlabel("recall")
plt.ylabel("precision")
plt.legend()
plt.show()

```

```python
# unixcoder recall
# precision vs recall
df = pd.read_csv('outputAdditional.csv')
df['recall'] = df.apply(lambda x: x['foundUnixcoder'] / x['potential'], axis=1)
df['precision'] = df.apply(lambda x: x['foundUnixcoder'] / (x['foundUnixcoder'] + x['unrelatedUnixcoder']), axis=1)
for val in dropAmounts:
    dfDrop = df[df.dropAmount == val]
    plt.plot(dfDrop['recall'], dfDrop['precision'], label="amount dropped: " + str(val))
print(max(df['recall'] * df['precision'] * 2 / (df['recall'] + df['precision'])))
ax = plt.gca()
ax.set_ylim(0,1)
plt.xlabel("recall")
plt.ylabel("precision")
plt.legend()
plt.show()



```
